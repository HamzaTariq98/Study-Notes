```python
tag_to_label = {'O': 0, 'B-PER': 1, 'I-PER': 2, 'B-ORG': 3, 'I-ORG': 4, 'B-LOC': 5, 'I-LOC': 6, 'B-MISC': 7, 'I-MISC': 8}
label_to_tag = {v:k for k,v in tag_to_label.items()}
```

```python
data = load_dataset("conll2003",trust_remote_code=True)
data = data.remove_columns(['id','pos_tags','chunk_tags'])
data = data.rename_column('tokens','words')
data = data.rename_column('ner_tags','label')
# data = data.map(lambda x:{'label_name':[[label_to_tag[tag] for tag in sample] for sample in x['label']]},batched = True)
```

```python
from transformers import AutoTokenizer
checkpoint = 'bert-base-cased'

tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
result = tokenizer(sample['words'],is_split_into_words=True)
result.tokens(),result
```

```python
def tokenize(sample):
  result = tokenizer(sample['words'],truncation = True,is_split_into_words=True)
  result['word_ids'] = result.word_ids()
  return result

data = data.map(tokenize)
```

```python
label_to_tag
```

```python
sample = data['train'][0]

def update_labels(sample):
  labels = []
  prev_id = 0
  for i,id in enumerate(sample['word_ids']):
    if id == None:
      labels.append(-100)
    elif id == prev_id:
      if sample['label'][prev_id] in [1,3,5,7]:
        labels.append(sample['label'][id]+1)
      else:
        labels.append(sample['label'][id])
    else:
      labels.append(sample['label'][id])
    prev_id = id
  return {'labels':labels}


data = data.map(update_labels)
```

```python
from transformers import AutoModelForTokenClassification, Trainer, TrainingArguments, DataCollatorForTokenClassification
from torch.optim import Adam
from sklearn.metrics import accuracy_score, f1_score



def compute_metrics(eval_preds):
  logits, labels = eval_preds
  predictions = logits.argmax(-1)
  true_labels = [[label_to_tag[i.item()] for i in sample if i !=-100]for sample in labels]
  true_predictions = [[label_to_tag[p.item()] for (i,p) in zip(label,preds) if i !=-100] for (label,preds) in zip(labels,predictions)]
  all_metrics = metric.compute(predictions = true_predictions,references = true_labels)
  acc = all_metrics['overall_accuracy']
  f1 = all_metrics['overall_f1']
  return{'acc':acc,'f1':f1}



model = AutoModelForTokenClassification.from_pretrained(checkpoint,id2label = label_to_tag,label2id = tag_to_label )
data_collator = DataCollatorForTokenClassification(tokenizer=tokenizer)
training_args = TrainingArguments("NER_EN_Token_Classification",
                                  evaluation_strategy = 'epoch',
                                  num_train_epochs = 5,
                                  per_device_train_batch_size = 20,
                                  per_device_eval_batch_size = 20,
                                  push_to_hub = True)



trainer = Trainer(model,
                  training_args,
                  data_collator = data_collator,
                  train_dataset = data_model['train'],
                  eval_dataset = data_model['test'],
                  tokenizer = tokenizer,
                  compute_metrics = compute_metrics
                  )

```

```python
trainer.train()
```
