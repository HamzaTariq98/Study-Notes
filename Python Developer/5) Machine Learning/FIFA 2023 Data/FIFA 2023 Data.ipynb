```python
import pandas as pd
import numpy as np
import seaborn as sns
import os,sys,pdb
from pathlib import Path



def clean_data(value):
    if value.find('K') != -1:
        return float(value.replace('€','').replace('K',''))*1000
    elif value.find('M') != -1:
        return float(value.replace('€','').replace('M',''))*1000000
    else:
        return float(value.replace('€',''))

files = [f"{os.getcwd()}\\Data\\{file}" for file in os.listdir(os.getcwd()+f"\Data")]
for file in files:
    print(file)



df = pd.read_csv(files[4].replace('\\',"/"))
df = df[['Name','Value','Wage']]




df['Value']= df['Value'].apply(clean_data)
df['Wage']= df['Wage'].apply(clean_data)
df = df[df['Value']>0]

df['Wage to Value ratio'] = df['Wage']/df['Value']

df.dropna(inplace= True)

df.sort_values('Wage to Value ratio', ascending= True)
```
