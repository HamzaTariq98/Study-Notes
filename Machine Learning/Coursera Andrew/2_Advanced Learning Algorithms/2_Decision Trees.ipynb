#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

X = np.array([['Pointy','Round','Present'],
             ['Floppy','Not round','Present'],
             ['Floppy','Round','Absent'],
             ['Pointy','Not round','Present'],
             ['Pointy','Round','Present'],
             ['Pointy','Round','Absent'],
             ['Floppy','Not round','Absent'],
             ['Pointy','Round','Absent'],
             ['Floppy','Round','Absent'],
             ['Floppy','Round','Absent']])
y = np.array([1,1,0,0,1,1,0,1,0,0])
X = pd.get_dummies(pd.DataFrame(X,columns = ['Ear Shape','Face Shape','Whiskers']),drop_first=True)
model = DecisionTreeClassifier()
model.fit(X,y)

plt.figure(figsize=(10, 8))
plot_tree(model, feature_names=list(X.columns), class_names=['Cat', 'Dog'], filled=True, rounded=True)
plt.show()
```

#### Random Forest, XGBoost

```python
from xgboost import XGBClassifier
from sklearn.datasets import load_digits
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier


X,y = load_digits(return_X_y=True)
model_xg = XGBClassifier(n_estimators = 20)
model_xg.fit(X,y)

model = RandomForestClassifier(n_estimators=20)
model.fit(X,y)

print('XGBoost acc',accuracy_score(y,model_xg.predict(X)))
print('Rand for acc',accuracy_score(y,model.predict(X)))
```
