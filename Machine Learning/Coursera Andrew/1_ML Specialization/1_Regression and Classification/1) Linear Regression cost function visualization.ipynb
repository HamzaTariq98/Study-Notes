```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook

class Linear_Model():


    def __init__(self,x,y):
        self.x = x
        self.y = y


    def cost_function(self,w,b=0):
        return (((self.y-(w*self.x+b))**2).sum())/2*len(self.x)


    def interpret_cost(self,W,B=None):
        if B is None:
            cost = np.array([])
            for _,w in enumerate(W):
                cost = np.append(cost,self.cost_function(w))

            plt.figure(figsize=(10,5))
            plt.plot(W,cost)
            plt.title('Cost function with different w values')
            plt.xlabel('w coeff')
            plt.ylabel('Cost values')
            plt.grid()
            plt.show()
        else:
            length = len(W)
            W,B  = np.meshgrid(W,B)
            Cost = np.ones((length,length))
            
            for i in range(length):
                for j in range(length):
                    w = W[i][j]
                    b = B[i][j]
                    Cost[i][j] = self.cost_function(w,b)

            fig = plt.figure(figsize=(15,5))
            ax = plt.axes(projection='3d')
            ax.plot_surface(W, B, Cost)
            ax.set_xlabel('W coef')
            ax.set_ylabel('B coef')
            ax.set_zlabel('Cost')
            ax.set_title('Cost Function')


x = np.array([1,2,8])
y = np.array([1,2,3])
W = np.linspace(-4,5,100)
B = np.linspace(-4,4,100)

model = Linear_Model(x,y)
```

```python
# Cost function with w param only
model.interpret_cost(W)
```

```python
# Cost function with w,b params
model.interpret_cost(W,B)
plt.show()
```
