```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

class Cluster_Model:
    def __init__(self, n_clusters):
        if n_clusters < 2:
            raise ValueError(f'n_clusters ({n_clusters}) must be greater or equal to 2')
        self.n_clusters = n_clusters
    
    
    def fit(self,X,y):
        self.X = X
        self.y = y
        self.n,self.m = X.shape
        if self.n_clusters > self.n:
            raise ValueError(f'choose n_clusters ({self.n_clusters}) to be less than or equal to X data points ({self.n})')
        
        
        # run clustering multiple times and store the centroid, cost of each trial
        centroids_values = []
        centroids_costs = []
        for i in range(10):
            cost,value  = self.initiate_algorithm(self)
            centroids_costs.append(cost)
            centroids_values.append(value)

            
        # Setting model centroid with the one that got minimum cost !
        self.centroids = centroids_values[np.argmin(centroids_costs,axis=0)]
        self.clusters = self.update_clusters(self)
        self.cost = self.cost_function(self)



    
    
    def predict(self,X):
        n = X.shape[0]
        distances = np.zeros((n,self.n_clusters))
        for i,x in enumerate(X):
            for j,centroid in enumerate(self.centroids):
                distances[i][j] = (((x-centroid)**2).sum())**0.5
        return np.argmin(distances, axis=1) 
    
    
    @staticmethod
    def initiate_algorithm(self):
        # initiate random centroids, clusters points
        random_indexes = np.random.choice(np.arange(self.n),size = self.n_clusters, replace = False)
        self.centroids = self.X[random_indexes]
        self.clusters = self.update_clusters(self)
        
        # update centroid values
        while(True):
            self.cost = self.cost_function(self)
            self.centroids = self.update_centroids(self)
            self.clusters = self.update_clusters(self)
            if self.cost == self.cost_function(self):
                break

        return self.cost, self.centroids
    
    
    @staticmethod
    def update_clusters(self):
        distances = np.zeros((self.n,self.n_clusters))
        for i,x in enumerate(self.X):
            for j,centroid in enumerate(self.centroids):
                distances[i][j] = (((x-centroid)**2).sum())**0.5
        return np.argmin(distances, axis=1)   
    
    
    @staticmethod
    def update_centroids(self):
        updated_centroids = np.zeros((self.n_clusters,self.m))
        for i in range(self.n_clusters):
            mask = self.clusters == i
            if np.any(X[mask]): # avoid incrementing empty centroid with no data assosiated with it
                updated_centroids[i] = self.X[mask].mean(axis = 0)
            
        return updated_centroids
    
    
    @staticmethod
    def cost_function(self):
        cost = 0
        for i,centroid in enumerate(self.centroids):
            mask = self.clusters == i
            filtered_X = X[mask]
            cost += ((filtered_X-centroid)**2).sum()
        return cost
    
 
        
X,y,centroids =  make_blobs(n_samples=200,n_features=2,cluster_std=2, centers=5, return_centers=True)
model = Cluster_Model(n_clusters=5)
sk_model = KMeans(n_clusters=5)

model.fit(X,y)
sk_model.fit(X,y)

y_pred = model.predict(X)

plt.figure(figsize=(10,5))
plt.scatter(x = X.T[0],y = X.T[1], c= y_pred)
plt.scatter(x=sk_model.cluster_centers_[:, 0], y=sk_model.cluster_centers_[:, 1], marker='X', s=200, c='k', label='sklearn model centroids')
plt.scatter(x=model.centroids[:, 0], y=model.centroids[:, 1], marker='o', s=100, c='red', label='manually built model centroids')
plt.legend()
plt.show()
```
