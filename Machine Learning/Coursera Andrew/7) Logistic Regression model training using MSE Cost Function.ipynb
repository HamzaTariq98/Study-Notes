```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(10)
noise = 5*np.random.randn(20)

x1 = np.arange(0,10,0.5)
x2 = x1**2

x = np.array([x1,x2])
y = x1**2 + x1+ noise
```

```python
def cost_function(w,b):
    return (((np.dot(x,w)+b-y)**2).sum())/m


def gradient_decent(w,b):
    temp_w = w

    for j in range(n):
        w[j] = temp_w[j] - alpha *(((np.dot(x,temp_w)+b-y)*x.T[0]).sum())/m
    b = b - alpha *(((np.dot(x,temp_w)+b-y)).sum())/m
    return w,b




def train_model(w,b):
    global iterations
    global cost
    w,b = gradient_decent(w,b)
    cost = np.append(cost,cost_function(w,b))
    iteration = 1
    iterations = np.append(iterations,iteration)

    while (abs(cost[-2]-cost_function(w,b))>0.000000001): # Keep running until get only less of 0.000000001 improvement!
        w,b = gradient_decent(w,b)
        cost = np.append(cost,cost_function(w,b))
        iteration += 1
        iterations = np.append(iterations,iteration)





def interpret_results():
    plt.figure(figsize=(10,5))
    xlabel = list(range(m))
    y_hat = np.dot(x,w)+b

    plt.scatter(xlabel,y, label = 'Actual data')
    plt.plot(xlabel,y_hat, label = 'Model Fit', c = 'k')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Quadratic Fit Curve using feature engineering x2 = x1^2')
    plt.legend()
    plt.grid()
    plt.show()
        
    plt.figure(figsize=(10,5))
    plt.plot(iterations,cost, label = 'Cost Function')
    plt.xlabel('Iteration')
    plt.ylabel('Cost function')
    plt.title('Learning Curve')
    plt.legend()
    plt.grid()
    plt.show()

x = np.column_stack((x1,x2))
m = len(x)
n = len(x[0])
alpha = 0.005
w = np.ones(n)
b = 1
iterations = np.array([1])
cost = np.array([cost_function(w,b)])
train_model(w,b)
interpret_results()

```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)

print(f'sklearn model params ==> {model.coef_} {model.intercept_}')
print(f'Gradient Decent parm ==> {w[::-1]} {b}  // {iterations[-1]} iterations')
```
