##### Python built in array  (List) appending follows a linear time
##### To get a Sequence with more efficient in adding new items we can use (Lined Lists)
 

Singly Liked List: each node refer for the node in front of it

Doubly Linked List: each node refer for the nodes in front and behind it

```python
class Node:
    """
    An object to store single node for Single Linked list
    
    Model attribute:-
    1) Data
    2) Link for next node in the list
    
    """
    data = None
    next_node = None
    
    def __init__(self,data):
        self.data = data
        
    def __repr__(self):
        return f"<Node data: {self.data}>"
    
    
class LinkedList:
    def __init__(self):
        self.head = None
        
        
    def is_empty(self):
        return self.head == None
    
    
    def size(self):
        node = self.head
        Linked_List_lenght = 0
        
        while node != None:
            Linked_List_lenght += 1
            node = node.next_node
        return Linked_List_lenght
    
    def prepend(self, data):        
        """
        Add an element a head of a Linked List ---
        node: The new node we want to add in the head of the LinkedList
        """
        node = Node(data)
        node.next_node = self.head
        self.head = node
        
    def insert(self,data,index):
        if index-1 >= self.size(): raise AssertionError(f"index({index}) value is outside LinkedList Range")
        if index ==0: return self.prepend(data)
        left_node = self.head
        count = 0
        while count <= index-2:
            count +=1
            left_node = left_node.next_node
        right_node = left_node.next_node

        left_node.next_node = Node(data)
        left_node.next_node.next_node = right_node 
        
    def search(self,data):
        current  = self.head
        while current is not None:
            if data == current.data:
                return current
            else:
                current = current.next_node
        return -1
    
    
    
    
    def __repr__(self):
        """
        Return a string representation of the linked list.
        """
        nodes = []
        current = self.head
        
        while current != None:
            nodes.append(str(current))
            current = current.next_node
        
        return " -> \n".join(nodes)

        


L1 = LinkedList()



L1.prepend(5)
L1.prepend(4)
L1.prepend(3)
L1.prepend(2)
L1.prepend(1)
L1.prepend(0)

L1.insert(6,6)
L1.insert(7,7)
L1.insert(8,8)
L1
```
