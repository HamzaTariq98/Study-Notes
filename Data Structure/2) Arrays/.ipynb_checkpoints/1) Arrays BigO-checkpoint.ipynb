**Static Arrays:** Their size are fixed
**Dynamic Arrays:** Their size expands when adding more elements

```python
import time,pdb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def list_append(lst):
    start_time = time.time()
    lst.append(5)
    return  time.time() - start_time

def list_insert(lst):
    start_time = time.time()
    lst.insert(0,5)
    return  time.time() - start_time


def list_remove(lst):
    start_time = time.time()
    lst.pop(0)
    return  time.time() - start_time

def run_trials(sample_size_list,func):
    result = []
    for size in sample_size_list:
        my_list = list(np.random.randint(0,10,size=(size)))
        fnc_time = func(my_list)
        result.append({'size':size,'fnc_time':fnc_time})
        
    result = pd.DataFrame(result)
    plt.figure(figsize=(10,5))
    plt.scatter(result['size'],result['fnc_time'])
    plt.title(f'{func.__name__}')
    plt.xlabel('Array Size')
    plt.ylabel('Time s')
    plt.show()
```

##### Appending elements to an array

```python
maximum = 2000000
step = 100000
sample_size_list = range(1,maximum,step)
run_trials(sample_size_list,list_append)
```

##### Insert element to an array

```python
import time,pdb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


maximum = 10000000
step = 1000000
sample_size_list = range(1,maximum,step)
run_trials(sample_size_list, list_insert)
```

##### Remove element froman array

```python
import time,pdb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


maximum = 10000000
step = 1000000
sample_size_list = range(1,maximum,step)
run_trials(sample_size_list, list_remove)
```
