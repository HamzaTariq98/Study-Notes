```python
from _1_Bubble_Sort import sort_bubble
from _2_Selection_Sort import sort_selection
from _3_Insertion_Sort import sort_insertion
from _4_Merge_Sort import sort_merge

import time

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def run_sort(arr, func):
    start_time = time.time()
    func(arr) 
    return time.time() - start_time


def show_results(results):
    results = pd.DataFrame(results)
    results_melted = pd.melt(results, id_vars=['Size'], var_name='Column', value_name='Value')

    sns.set(style="whitegrid") 
    plt.figure(figsize=(10, 6)) 
    sns.lineplot(data=results_melted, x='Size', y='Value', hue='Column')
    plt.title("Time Complexity for variety of sorting methods")
    plt.xlabel("Size")
    plt.ylabel("Value")
    plt.legend(title='Legend')
    plt.show()

results = []
for size in range(10,2500,100):
    arr = np.random.randint(-9999,9999,size= size)
    t1 = run_sort(arr,sort_bubble)
    t2 = run_sort(arr,sort_selection)
    t3 = run_sort(arr,sort_insertion)
    t4 = run_sort(arr,sort_merge)
    t5 = run_sort(arr,sorted)
    results.append({
        'Size': size,
        'Bubble': t1,
        'Selection': t2,
        'Insertion': t3,
        'Merge': t4,
        'Python Built in':t5
    })



show_results(results)

```
